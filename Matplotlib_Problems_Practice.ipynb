````python
import pandas as pd
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales  = [200, 220, 250, 280, 300, 330]

plt.plot(months,sales , color= "purple", linestyle = "--", marker="o")

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True, alpha = 0.3)


plt.show()
````

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
product_A = [200, 220, 250, 280, 300, 330]
product_B = [150, 180, 160, 210, 240, 260]
product_C = [100, 120, 140, 130, 160, 190]

plt.plot(months , product_A , color= "orange" , marker = "o", linestyle="--", label="Product A")
plt.plot(months , product_B , color= "black" , marker = "o", linestyle="-", label="Product B")
plt.plot(months , product_C , color= "green" , marker = "o", linestyle=":", label="Product C")

plt.title("Monthly Sales on Products")
plt.xlabel("Months", fontsize=12)
plt.ylabel("Sales",fontsize=12)
plt.legend()
plt.grid(True,alpha=0.3)

plt.show()
````

````python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df

plt.hist(df["Age"], bins=30,color="orange", edgecolor="black")

plt.title("Age distribution of titanic Passengers", fontsize = 13)
plt.xlabel("Age")
plt.ylabel("No. Passengers")
plt.grid(axis = "y",alpha=0.3)

plt.show()
````

````python
# colors = df["Survived"].map({0:"red", 1:"green"})

plt.scatter(df['Age'], df['Fare'], c="orange", edgecolor="black", linewidth=0.4,alpha=0.3)

plt.title("Age vs Fare", fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True, alpha=0.3)

plt.show()
````

````python
# With color coding in scatter plots 

colors = df["Survived"].map({0:"maroon", 1:"seagreen"})

plt.scatter(df['Age'], df['Fare'], c=colors, edgecolor="black", linewidth=0.4)

# Manual legend
import matplotlib.patches as mpatches
legend_elements = [
    mpatches.Patch(color="maroon",     label="Did not survive"),
    mpatches.Patch(color="seagreen", label="Survived")
]

plt.legend(handles=legend_elements, fontsize=11)

plt.title("Age vs Fare", fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True, alpha=0.3)

plt.show()
````

PIE_CHART

````python
filter_survive = df["Pclass"].value_counts()
labels = ["Third Class", "First Class","Second Class"]
colors = ["maroon", "seagreen","orange"]
explode = (0.01, 0.02, 0.03)

plt.pie(filter_survive , labels = labels, colors = colors , explode=explode)

plt.title("Survival rate on Titanic", fontweight="bold")
plt.show()
````

````python
plt.figure(figsize=(14, 10))
plt.suptitle("Titanic Data Analysis Dashboard",
             fontsize=18, fontweight="bold")

# Chart 1 — top left
plt.subplot(2, 2, 1)                # (rows, cols, position)
plt.hist(df["Age"].dropna(), bins=30,
         color="steelblue", edgecolor="black", alpha=0.7)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(alpha=0.3)

# Chart 2 — top right
plt.subplot(2, 2, 2)
survived = df["Survived"].value_counts()
plt.bar(["Did not survive", "Survived"],
        survived.values,
        color=["coral", "steelblue"],
        edgecolor="black")
plt.title("Survival Count")
plt.ylabel("Count")
plt.grid(axis="y", alpha=0.3)

# Chart 3 — bottom left
plt.subplot(2, 2, 3)
pclass_counts = df["Pclass"].value_counts().sort_index()
plt.bar(["Class 1", "Class 2", "Class 3"],
        pclass_counts.values,
        color=["gold", "silver", "brown"],
        edgecolor="black")
plt.title("Passengers per Class")
plt.ylabel("Count")
plt.grid(axis="y", alpha=0.3)

# Chart 4 — bottom right
plt.subplot(2, 2, 4)
plt.scatter(df["Age"], df["Fare"],
            alpha=0.4, color="steelblue",
            edgecolor="black", linewidth=0.3)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

````

````python

# ## 📝 Practice Tasks — Matplotlib Only

# Use the Titanic dataset for all tasks:
# ```
# 1. Line plot — show passenger count 
#    per Pclass as a trend line
#    (x = class, y = count)
pclass = df["Pclass"].value_counts().sort_index()
# print(pclass)
plt.figure(figsize=(10,6))

plt.plot(pclass.index , pclass.values , color="orange", linestyle="--",marker="o")
plt.title("Passenger Count Analysis" ,fontsize = 14, fontweight="bold")
plt.xlabel("Pclass", fontsize = 12)
plt.ylabel("Count of Passengers",fontsize = 12)
plt.grid(True, alpha =0.4)
plt.xticks([1, 2, 3],
           ["Class 1", "Class 2", "Class 3"])

plt.tight_layout()
plt.show()



# 📊 INSIGHT 1: Class 3 has most passengers (491) —
#               more than Class 1 and 2 combined
# 📊 INSIGHT 2: Class 2 has fewest passengers (184) —
#               middle class was least represented
# 📊 INSIGHT 3: Sharp rise from Class 2 to Class 3
#               shows Titanic was primarily a ship
#               for working class passengers
````

````python
# 2. Bar chart — passenger count 
#    per Pclass with value labels on top
pclass = df["Pclass"].value_counts().sort_index()
colors_3 = ["orange", "maroon", "seagreen"]
labels_values = ["Class-1","Class-2", "Class-3"]

plt.figure(figsize=(10,6))

for x,y, c , labels in zip(pclass.index , pclass.values, colors_3, labels_values):
    plt.bar(x,y, color = c , edgecolor='black' ,label=labels)

plt.title("Passenger Count Analysis" ,fontsize = 14, fontweight="bold")
plt.xlabel("Pclass", fontsize = 12)
plt.ylabel("Count of Passengers",fontsize = 12)
plt.grid(True, alpha =0.4)
plt.xticks([1, 2, 3],
           ["Class 1", "Class 2", "Class 3"])

plt.tight_layout()
plt.legend()

plt.show()


# 📊 INSIGHT 1: Class 3 has most passengers (491) —
#               more than Class 1 and 2 combined
# 📊 INSIGHT 2: Class 2 has fewest passengers (184) —
#               middle class was least represented
# 📊 INSIGHT 3: Sharp rise from Class 2 to Class 3
#               shows Titanic was primarily a ship
#               for working class passengers
````

````python
# 3. Horizontal bar chart — 
#    average Fare per Pclass
pclass = df["Pclass"].value_counts().sort_index()
colors_3 = ["orange", "maroon", "seagreen"]
labels_values = ["Class-1","Class-2", "Class-3"]

plt.figure(figsize=(10,6))

for x,y, c , labels in zip(pclass.index , pclass.values, colors_3, labels_values):
    plt.barh(x,y, color = c , edgecolor='black' ,label=labels)

plt.title("Passenger Count Analysis" ,fontsize = 14, fontweight="bold")
plt.xlabel("Count of Passengers", fontsize = 12)
plt.ylabel("Pclass",fontsize = 12)
plt.grid(axis="x", alpha =0.4)
plt.yticks([1, 2, 3],
           ["Class 1", "Class 2", "Class 3"])

plt.tight_layout()
plt.legend()

plt.show()

````
